```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
def is_prime(x):
    end = int(math.sqrt(x))
    for i in range(2, end+1):
        if x % i == 0:
            return False
    return True
```

```python
def xstep(x, flip=False):
    if flip == True:
        return x - np.array([1, 0])
    return x + np.array([1, 0])

def ystep(x, flip=False):
    if flip == True:
        return x - np.array([0, 1])
    return x + np.array([0, 1])

n = [np.array([0, 0])]

def root(n, i, flip):
    for j in range(i):
        n.append(xstep(n[-1], flip))

    for j in range(i):
        n.append(ystep(n[-1], flip))

    return n

l = []
for i in range(120):
    flip = False
    if i == 0:
        k = root(n, i, flip)
        l += k
    else: 
        if i % 2 == 0:
            flip = True
        k = root([k[-1]], i, flip)
        k.pop(0)
        l += k
```

```python
%matplotlib inline
primes = []
for j, i in enumerate(l):
    j += 1
    if is_prime(j) == True:
        primes.append(i)
        
l = np.array(l)

xs = [i[0] for i in l]
ys = [i[1] for i in l]
pxs = [i[0] for i in primes]
pys = [i[1] for i in primes]

plt.figure(figsize=(18,18))
plt.scatter(pxs, pys, color='orange', marker="D", zorder=5, s=42)

plt.plot(xs, ys, '.', color='slategray', alpha=1, zorder=3)

```
